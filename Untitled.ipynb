```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

#Formatter to show dollar amounts properly
money_fmt = mticker.FuncFormatter(lambda x, _: f"${x:,.0f}")

print("Ready!")
```

```python
df = pd.read_excel("task 2/finance_transactions_dataset.xlsx")

#Quick check
print(df.shape)  #should be around (1000, 14)
print(df.columns.tolist())
```

```python
#3a - Preview first rows
df.head()

#3b - Column types and missing counts
df.info()

#3c - Summary statistics
df.describe().round(2)

#3d - Count missing values
df.isnull().sum()
```

```python
#4a - Remove rows with missing critical values
df.dropna(subset=['Amount', 'Net_Amount', 'Transaction_Date', 'Approval_Status'], inplace=True)
print("Rows remaining:", len(df))
```

```python
#4b - Fill gaps in text columns
df['Company'] = df['Company'].fillna('Unknown')
df['Department'] = df['Department'].fillna('Unknown')
df['Analyst'] = df['Analyst'].fillna('Unknown')
df['Category'] = df['Category'].fillna('N/A')
```

```python
#4c - Convert text dates to proper datetime
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], errors='coerce')
df.dropna(subset=['Transaction_Date'], inplace=True)

#Extract Year and Month
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.to_period('M').astype(str)

print("Sample dates:")
print(df[['Transaction_Date', 'Year', 'Month']].head(3))
```

```python
#4d - Remove duplicates
df.drop_duplicates(inplace=True)
print("Final row count:", len(df))
```

```python
#5a - Bar chart

df.groupby('Transaction_Type')['Amount'].sum().sort_values().plot(kind='bar', color='steelblue')

plt.title('Total Amount by Transaction Type')
plt.ylabel('Total Amount ($)')
plt.xticks(rotation=15)
plt.gca().yaxis.set_major_formatter(money_fmt) # show readable dollar amounts
plt.tight_layout()
plt.show()
```

```python
#5b - Grouped bar chart
df.groupby('Sector')[['Net_Amount', 'Tax_Amount']].sum().plot(kind="bar", color=['steelblue', 'coral'])

plt.title('Net Amount vs Tax Amount by Sector')
plt.ylabel('Amount($)')
plt.xticks(rotation=20)
plt.gca().yaxis.set_major_formatter(money_fmt)
plt.tight_layout()
plt.show()
```

```python
#5c - line chart
df.groupby('Month')["Amount"].sum().plot(kind='line', marker='o', color='steelblue')

plt.title('Monthly Transaction Volume')
plt.ylabel('Total Amount($)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(money_fmt)
plt.show()
```

```python
#5d - pie chart
df['Approval_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Approval Status Distribution')
plt.ylabel('') #hides default label
plt.tight_layout()
plt.show()
```

```python
#5e - horinzontal bar chart
df.groupby('Analyst')['Amount'].sum().nlargest(5).plot(kind='barh', color='mediumseagreen')

plt.title('Top 5 Analysts by Total Amount')
plt.xlabel('Total_Amount ($)')
plt.gca().xaxis.set_major_formatter(money_fmt)
plt.show()
```
